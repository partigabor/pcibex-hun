```python
import pandas as pd

et_path = r"c:\Users\parti\Projects\pcibex-hun\et_data.csv"
et_df = pd.read_csv(et_path)

dwell_stats = (
    et_df.groupby(["condition", "chosen_type"])["total_dwell"]
    .agg(total_dwell_mean="mean", total_dwell_sd="std")
    .reset_index()
)

print(dwell_stats)
```

```python
from scipy.stats import ttest_ind

contrastive = et_df[et_df["condition"] == "contrastive"]
a_dwell = contrastive.loc[contrastive["chosen_type"] == "A", "total_dwell"]
b_dwell = contrastive.loc[contrastive["chosen_type"] == "B", "total_dwell"]

t_stat, p_val = ttest_ind(a_dwell, b_dwell, equal_var=False, nan_policy="omit")

print(f"A count={a_dwell.size}, mean={a_dwell.mean():.2f}, sd={a_dwell.std(ddof=1):.2f}")
print(f"B count={b_dwell.size}, mean={b_dwell.mean():.2f}, sd={b_dwell.std(ddof=1):.2f}")
print(f"Welch t-test: t={t_stat:.3f}, p={p_val:.4f}")
```
